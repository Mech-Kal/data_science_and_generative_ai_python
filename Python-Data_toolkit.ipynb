```python
# 1. Create a scatter plot to visualize the relationship between two variables, by generating a synthetic dataset.

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Synthetic dataset
np.random.seed(1)
x = np.random.rand(100)
y = x + np.random.normal(0, 0.1, size=100)

df_scatter = pd.DataFrame({'X': x, 'Y': y})

sns.scatterplot(data=df_scatter, x='X', y='Y')
plt.title("Scatter Plot - Synthetic Data")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

```python
# 5. Generate a synthetic dataset with correlated features.
# Visualize the correlation matrix of a dataset using a heatmap.

# Create synthetic dataset
np.random.seed(10)
df_corr = pd.DataFrame({
    'Feature A': np.random.rand(100),
    'Feature B': np.random.rand(100),
    'Feature C': np.random.rand(100),
})

# Introduce correlation
df_corr['Feature D'] = df_corr['Feature A'] * 0.5 + df_corr['Feature B'] * 0.5

# Compute correlation matrix
corr_matrix = df_corr.corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# 1. Generate a 3D scatter plot to visualize data points in a three-dimensional space.

import plotly.express as px
import numpy as np
import pandas as pd

np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 18, 300),
    'Y': np.random.uniform(-10, 18, 300),
    'Z': np.random.uniform(-16, 18, 300)
}
df = pd.DataFrame(data)

fig = px.scatter_3d(df, x='X', y='Y', z='Z', title='3D Scatter Plot')
fig.show()
```

```python
# 2. Create a violin plot to display the distribution of scores across grade categories.

np.random.seed(15)
data = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}
df = pd.DataFrame(data)

fig = px.violin(df, x='Grade', y='Score', box=True, points='all', title='Violin Plot of Scores by Grade')
fig.show()
```

```python
# 3. Generate a heatmap to visualize the variation in sales across different months and days.

np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 180),
    'Day': np.random.choice(range(1, 31), 180),
    'Sales': np.random.randint(1000, 5000, 180)
}
df = pd.DataFrame(data)

heatmap_data = df.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')

import plotly.figure_factory as ff
fig = ff.create_annotated_heatmap(
    z=heatmap_data.values,
    x=heatmap_data.columns.tolist(),
    y=heatmap_data.index.tolist(),
    colorscale='Viridis'
)
fig.update_layout(title='Sales Heatmap: Days vs Months')
fig.show()
```

```python
# 4. Generate a 3D surface plot to visualize the function z = sin(sqrt(x^2 + y^2))

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

fig = px.surface(z=z, x=np.linspace(-5, 5, 100), y=np.linspace(-5, 5, 100), title='3D Surface Plot')
fig.show()
```

```python
# 1. Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
from bokeh.plotting import figure, show, output_file
import numpy as np

output_file("sine_wave.html")
x = np.linspace(0, 10, 500)
y = np.sin(x)

p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')
p.line(x, y, line_width=2)

show(p)
```

```python
# 2. Create a Bokeh scatter plot using randomly generated x and y values.
# Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.
from bokeh.plotting import figure, show, output_file
from bokeh.palettes import Viridis256
from bokeh.models import ColumnDataSource
import numpy as np

output_file("scatter_plot.html")

x = np.random.rand(100)
y = np.random.rand(100)
sizes = np.random.randint(10, 30, size=100)
colors = np.random.choice(Viridis256, size=100)

source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

p = figure(title="Random Scatter Plot", x_axis_label='x', y_axis_label='y')
p.scatter('x', 'y', size='sizes', color='colors', alpha=0.6, source=source)

show(p)
```

```python
# 3. Generate a Bokeh bar chart representing the counts of different fruits using the given dataset.
from bokeh.plotting import figure, show, output_file
from bokeh.transform import dodge
from bokeh.models import ColumnDataSource

output_file("fruit_bar_chart.html")

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [208, 25, 30, 35]
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

p = figure(x_range=fruits, title="Fruit Counts", y_axis_label="Count", x_axis_label="Fruit")
p.vbar(x='fruits', top='counts', width=0.6, source=source, color="orange")

show(p)
```

```python
# 4. Create a Bokeh histogram to visualize the distribution of the given data.
from bokeh.plotting import figure, show, output_file
import numpy as np

output_file("histogram.html")

data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

p = figure(title="Histogram", x_axis_label='Value', y_axis_label='Frequency')
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.7)

show(p)
```

```python
# 5. Create a Bokeh heatmap using the provided dataset.
from bokeh.plotting import figure, show, output_file
from bokeh.models import LinearColorMapper, ColorBar
from bokeh.transform import transform
from bokeh.palettes import Viridis256
import numpy as np
import pandas as pd

output_file("heatmap.html")

data_heatmap = np.random.rand(18, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 18)
xx, yy = np.meshgrid(x, y)

df = pd.DataFrame({'x': xx.flatten(), 'y': yy.flatten(), 'value': data_heatmap.flatten()})
mapper = LinearColorMapper(palette=Viridis256, low=df.value.min(), high=df.value.max())

p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y')
p.rect(x="x", y="y", width=0.1, height=0.1, source=df,
       fill_color=transform('value', mapper), line_color=None)

color_bar = ColorBar(color_mapper=mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

show(p)
```
